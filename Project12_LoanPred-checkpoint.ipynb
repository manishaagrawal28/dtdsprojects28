```python
#removing outliers
from scipy.stats import zscore
z_score=abs(zscore(loan))
print(loan.shape)
loan_data=loan.loc[(z_score<3).all(axis=1)]
print(loan_data.shape)
#shape changed from 529,12 to 475,12
```

```python
#let separate input and output variables to be used for the problem
x=pd.DataFrame(loan_data.drop(columns=['Loan_Status']))
x=pd.DataFrame(loan_data.iloc[:,0:11])
y=pd.DataFrame(loan_data['Loan_Status'])
```

```python
#checking for skewness
x.skew()  #skewed data present
```

```python
#removing skewness
import numpy as np
for col in x.columns:
    if x.skew().loc[col]>0.55:
        x[col]=np.log1p(x[col])
```

```python
#again checking for skewness
x.skew()
```

```python
#Using standardscaler tecnique for scaling the data

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

For this dataset we can find the output by using Classification Algorithm.

```python
#splitting the data
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=9,test_size=0.3,stratify=y)
```

```python
#importing algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#creating the object
lr=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=6)
sv=SVC()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
gnb=GaussianNB()
```

```python
#Using GRidSearchCV for finding best parameters
from sklearn.model_selection import GridSearchCV
```

```python
#for knn:
neighbors={"n_neighbors":range(1,30)}
gknn=GridSearchCV(knn,neighbors,cv=10)
gknn.fit(x,y)
gknn.best_params_
```

```python
#for sv:
kernel={"kernel":["linear","rbf","poly"]}
gsv=GridSearchCV(sv,kernel,cv=10)
gsv.fit(x,y)
gsv.best_params_
```

```python
#for DcT:
grid_param={"random_state":range(0,20),"criterion":['gini','entropy']}
gdt=GridSearchCV(dt,param_grid=grid_param,scoring='accuracy',cv=10)
gdt.fit(x,y)
gdt.best_params_
```

```python
#for RandomForestClassifier
parameters={"n_estimators":[10,100,500,1000]}
clf=GridSearchCV(rf,parameters,cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
#providing parameters
lr=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=7)
sv=SVC(kernel="rbf")
dt=DecisionTreeClassifier(random_state=9,criterion="gini")
rf=RandomForestClassifier(n_estimators=500)
gnb=GaussianNB()
```

```python
models=[]
models.append(("LogisticRegression",lr))
models.append(("KNeighborsClassifier",knn))
models.append(("SVC",sv))
models.append(("DecisionTreeClassifier",dt))
models.append(("GaussianNB",gnb))
models.append(("RandomForestClassifier",rf))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
```

```python
Model=[]
score=[]
cvs=[]
roc_score=[]
for name,model in models:
    print('*********************',name,'***************************')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print("accuracy_score",AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring="accuracy").mean()
    print("cross_val_score=",sc)
    cvs.append(sc*100)
    false_positive_rate,true_positive_rate,threshold=roc_curve(y_test,pre)
    roc_auc=auc(false_positive_rate,true_positive_rate)
    print("roc_auc_score=",roc_auc)
    roc_score.append(roc_auc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,pre))
    print("\n")
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(false_positive_rate,true_positive_rate,label='AUC=%0.2f'%roc_auc)
    plt.plot([0,1],[0,1],'r--')
    plt.legend(loc='lower right')
    plt.ylabel('True positive rate')
    plt.xlabel('False positive rate')
    print('\n\n')
```

```python
result=pd.DataFrame({'Model':Model,'Accuracy_score':score,'Cross_Val_Score':cvs,'roc_auc_curve':roc_score})
result
```
