```python
#removing outliers
from scipy.stats import zscore
z_score=abs(zscore(heart))
print(heart.shape)
heart_data=heart.loc[(z_score<3).all(axis=1)]
print(heart_data.shape)
#shape changed from 302,14 to 293,14
```

```python
#let separate input and output variables to be used for the problem
x=heart_data.drop(columns=['num'])
x=pd.DataFrame(heart_data.iloc[:,0:13])
y=pd.DataFrame(heart_data['num'])
```

```python
#checking for skewness
x.skew()  #skewed data present
```

```python
#removing skewness with log
import numpy as np
for col in x.columns:
    if x.skew().loc[col]>0.55:
        x[col]=np.log1p(x[col])
```

```python
#again checking for skewness
x.skew()
```

```python
#scaling down using standardscaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#using classification algorithm for this dataset

from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=9,test_size=0.2,stratify=y)
```

```python
#importing algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#creating object file of the algorithm
lg=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=6)
sv=SVC()
dt=DecisionTreeClassifier(random_state=6)
rf=RandomForestClassifier()
```

```python
#by using gridsearch cv finding best parameters
#for knn:
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
neighbors={"n_neighbors":range(1,30)}
gknn=GridSearchCV(knn,neighbors,cv=10)
gknn.fit(x,y)
gknn.best_params_
```

```python
#by using gridsearch cv finding best parameters
#for SVC:
import warnings
warnings.filterwarnings('ignore')
kernel={"kernel":["linear","rbf","poly"]}
gsv=GridSearchCV(sv,kernel,cv=10)
gsv.fit(x,y)
gsv.best_params_
```

```python
#by using gridsearch cv finding best parameters
#for DCT:
import warnings
warnings.filterwarnings('ignore')
grid_param={"random_state":range(0,20),"criterion":['gini','entropy']}
gdt=GridSearchCV(dt,param_grid=grid_param,scoring='accuracy',cv=10)
gdt.fit(x,y)
gdt.best_params_
```

```python
#providing parameters
knn=KNeighborsClassifier(n_neighbors=17)
sv=SVC(kernel="linear")
dt=DecisionTreeClassifier(random_state=9,criterion="gini")
rf=RandomForestClassifier()
```

```python
models=[]
models.append(("LogisticRegression",lg))
models.append(("KNeighborsClassifier",knn))
models.append(("SVC",sv))
models.append(("DecisionTreeClassifier",dt))
models.append(("RandomForestClassifier",rf))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
Model=[]
score=[]
cvs=[]

for name,model in models:
    print('***************',name,'*******************')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pred=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pred)
    print("accuracy_score",AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring="accuracy").mean()
    print("cross_val_score=",sc)
    cvs.append(sc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,pred))
    print("\n")
    con=confusion_matrix(y_test,pred)
    print(con)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(con,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot([0,1],[0,1],'r--')
    print('\n\n')
```

```python
#Using AdaBoost Classifier
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(x_train, y_train)
ada_predict=ada.predict(x_test)
```

```python
ada_conf_matrix = confusion_matrix(y_test, ada_predict)
ada_acc_score = accuracy_score(y_test, ada_predict)
print(ada_conf_matrix)
print(ada_acc_score*100)
```

```python
#showing the results of all the models
result=pd.DataFrame({'Model':Model,'Accuracy_score':score,'Cross_Val_Score':cvs})
result
```
