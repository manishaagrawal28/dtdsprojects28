```python
#importing pandas library
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#load the dataset
ds=pd.read_csv("Redwine.csv")
ds
```

```python
#putting data into dataframe
wine=pd.DataFrame(data=ds)
```

```python
#details of statistical data of individual column
wine.describe()
```

```python
#checking for skewness

wine.skew() #skewed  data present
```

```python
#removing skewness with log

import numpy as np
for col in wine.columns:
    if wine.skew().loc[col]>0.55:
        wine[col]=np.log1p(wine[col])
```

```python
#again checking for skewness
wine.skew()
```

```python
#checking for outliers

from scipy.stats import zscore
z=np.abs(zscore(wine))
print(wine.shape)
winedata=wine.loc[(z<3).all(axis=1)]
print(winedata.shape)
#outliers removed .shape changed from 1599,12 to 1480,12
```

```python
#separating input and output

x=winedata.iloc[:,0:11]
y=winedata.iloc[:,11]
```

```python
#pca to reduce the no. of columns
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(x)
```

```python
pca_new = PCA(n_components=8)
x= pca_new.fit_transform(x)
```

```python
#using standard scaler for scaling the data
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
#dividing the data into training and testing data
from sklearn.model_selection import train_test_split,cross_val_score
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=9,test_size=0.2,stratify=y)
```

```python
print(train_x.shape,test_x.shape)
```

```python
print(train_y.shape,test_y.shape)
```

```python
#using classification algorithm for this dataset
#importing algorithms

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
#creating object of the algorithm
lg=LogisticRegression()
gnb=GaussianNB()
knn=KNeighborsClassifier(n_neighbors=6)
sv=SVC()
dt=DecisionTreeClassifier(random_state=6)
```

```python
#by using gridsearchcv finding best parameters
from sklearn.model_selection import GridSearchCV
neighbors={"n_neighbors":range(1,30)}
gknn=GridSearchCV(knn,neighbors,cv=10)
gknn.fit(x,y)
gknn.best_params_
```

```python
#for SVC:
kernel={"kernel":["linear","rbf","poly"]}
gsv=GridSearchCV(sv,kernel,cv=10)
gsv.fit(x,y)
gsv.best_params_
```

```python
#For DCT:
grid_param={"random_state":range(0,20),"criterion":['gini','entropy']}
gdt=GridSearchCV(dt,param_grid=grid_param,scoring='accuracy',cv=10)
gdt.fit(x,y)
gdt.best_params_
```

```python
#providing the parameters
knn=KNeighborsClassifier(n_neighbors=27)
sv=SVC(kernel="rbf")
dt=DecisionTreeClassifier(random_state=7,criterion="entropy")
```

```python
models=[]
models.append(("LogisticRegression",lg))
models.append(("GaussianNB",gnb))
models.append(('KNeighborsClassifier',knn))
models.append(('SVC',sv))
models.append(('DecisionTreeClassifier',dt))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
```

```python
Model=[]
score=[]
cvs=[]
for name,model in models:
    print('************',name,'************')
    print('\n')
    Model.append(name)
    model.fit(train_x,train_y)
    print(model)
    pre=model.predict(test_x)
    print('\n')
    AS=accuracy_score(test_y,pre)
    print('Accuracy_score = ',AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model, x, y,cv=10,scoring='accuracy').mean()
    print('Cross_Val_Score = ',sc)
    cvs.append(sc*100)
    print('\n')
    print('classification_report\n',classification_report(test_y,pre))
    print('\n')
    cm=confusion_matrix(test_y,pre)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot([0,1],[0,1],'r--')
    print('\n\n')
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(train_x, train_y)
rf_predict=rf.predict(test_x)
```

```python
rf_conf_matrix = confusion_matrix(test_y, rf_predict)
rf_acc_score = accuracy_score(test_y, rf_predict)
print(rf_conf_matrix)
print(rf_acc_score*100)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(train_x, train_y)
ada_predict=ada.predict(test_x)
```

```python
ada_conf_matrix = confusion_matrix(test_y, ada_predict)
ada_acc_score = accuracy_score(test_y, ada_predict)
print(ada_conf_matrix)
print(ada_acc_score*100)
```

```python
#showing the performance of all the models
result=pd.DataFrame({'Model':Model,'Accuracy_score':score,'Cross_val_score':cvs})
result
```
