```python
#removing outliers
from scipy.stats import zscore
z_score=abs(zscore(insurance))
print(insurance.shape)
insurancedata=insurance.loc[(z_score<3).all(axis=1)]
print(insurancedata.shape)
#final shape reduced to (1309,8) from (1338,8)
```

```python
#checking for skewness
insurancedata.skew()
#skewed data are present in column children,smoker and charges
```

```python
#removing skewness
import numpy as np
for col in insurancedata.columns:
    if insurancedata.skew().loc[col]>0.55:
        insurancedata[col]=np.log1p(insurancedata[col])
```

```python
#again checking for skewness
insurancedata.skew()
```

```python
#let separate input and output variables to be used for the problem
in_x=insurancedata.drop(columns=['insuranceclaim'])
y=pd.DataFrame(insurancedata['insuranceclaim'])
```

```python
#scaling down using standardscaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(in_x)
x=pd.DataFrame(x,columns=in_x.columns)
```

```python
#spilitting the data
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=9,test_size=0.2,stratify=y)
```

```python
#using classification algorithm for this dataset
#importing algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
#creating an object of the algorithm
lg=LogisticRegression()
gnb=GaussianNB()
knn=KNeighborsClassifier(n_neighbors=6)
sv=SVC()
dt=DecisionTreeClassifier(random_state=6)
```

```python
#by using gridsearch cv finding best parameters
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
neighbors={"n_neighbors":range(1,30)}
gknn=GridSearchCV(knn,neighbors,cv=10)
gknn.fit(x,y)
gknn.best_params_
```

```python
#by using gridsearch cv finding best parameters
kernel={"kernel":["linear","rbf","poly"]}
gsv=GridSearchCV(sv,kernel,cv=10)
gsv.fit(x,y)
gsv.best_params_
```

```python
#by using gridsearch cv finding best parameters
grid_param={"random_state":range(0,20),"criterion":['gini','entropy']}
gdt=GridSearchCV(dt,param_grid=grid_param,scoring='accuracy',cv=10)
gdt.fit(x,y)
gdt.best_params_
```

```python
#the best parameters are provided for hyperparameter tuning
knn=KNeighborsClassifier(n_neighbors=3)
sv=SVC(kernel="rbf")
dt=DecisionTreeClassifier(random_state=0,criterion="entropy")
```

```python
models=[]
models.append(("LogisticRegression",lg))
models.append(("GaussianNB",gnb))
models.append(("KNeighborsClassifier",knn))
models.append(("SVC",sv))
models.append(("DecisionTreeClassifier",dt))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
```

```python
Model=[]
score=[]
cvs=[]
roc_score=[]
for name,model in models:
    print('***************',name,'*******************')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pred=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pred)
    print("accuracy_score",AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring="accuracy").mean()
    print("cross_val_score=",sc)
    cvs.append(sc*100)
    false_positive_rate,true_positive_rate,threshold=roc_curve(y_test,pred)
    roc_auc=auc(false_positive_rate,true_positive_rate)
    print("roc_auc_score=",roc_auc)
    roc_score.append(roc_auc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,pred))
    print("\n")
    con=confusion_matrix(y_test,pred)
    print(con)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(con,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(false_positive_rate,true_positive_rate,label='AUC=%0.2f'%roc_auc)
    plt.plot([0,1],[0,1],'r--')
    plt.legend(loc='lower right')
    plt.ylabel('True positive rate')
    plt.xlabel('False positive rate')
    print('\n\n')
```

```python
#showing the results of all the models
result=pd.DataFrame({'Model':Model,'Accuracy_score':score,'Cross_Val_Score':cvs,'roc_auc_curve':roc_score})
result
```

DecisionTreeClassifier gives the best result.So this algorithm with the above hyperparameter should provide good predictions about whether the person will be using insurance claim.
