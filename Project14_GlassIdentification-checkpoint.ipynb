```python
#removing outliers
from scipy.stats import zscore
z_score=abs(zscore(glass))
print(glass.shape)
glass_identity=glass.loc[(z_score<3).all(axis=1)]
print(glass_identity.shape)
#shape changed from 214,10 to 194,10
```

```python
#let separate input and output variables to be used for the problem
x=glass_identity.drop(columns=['Type of glass'])
x=pd.DataFrame(glass_identity.iloc[:,0:9])
y=pd.DataFrame(glass_identity['Type of glass'])
```

```python
#checking for skewness
glass_identity.skew()  #skewed data present
```

```python
#removing skewness
import numpy as np
for col in glass_identity.columns:
    if glass_identity.skew().loc[col]>0.55:
        glass_identity[col]=np.log1p(glass_identity[col])
```

```python
#again checking for skewness
glass_identity.skew()
```

```python
#scaling down using standardscaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

Using Classification Algorithm for this dataset

```python
#splitting of the dataset
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=9,test_size=0.2,stratify=y)
```

```python
#importing algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
#creating the object
lg=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=6)
sv=SVC()
dt=DecisionTreeClassifier(random_state=6)
rf=RandomForestClassifier()
ada = AdaBoostClassifier()
```

```python
#by using gridsearch cv finding best parameters

import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
neighbors={"n_neighbors":range(1,30)}
gknn=GridSearchCV(knn,neighbors,cv=10)
gknn.fit(x,y)
gknn.best_params_
```

```python
#by using gridsearchcv finding best parameters for SVC:

import warnings
warnings.filterwarnings('ignore')
kernel={"kernel":["linear","rbf","poly"]}
gsv=GridSearchCV(sv,kernel,cv=10)
gsv.fit(x,y)
gsv.best_params_
```

```python
#For DCT Classifier:

import warnings
warnings.filterwarnings('ignore')
grid_param={"random_state":range(0,20),"criterion":['gini','entropy']}
gdt=GridSearchCV(dt,param_grid=grid_param,scoring='accuracy',cv=10)
gdt.fit(x,y)
gdt.best_params_
```

```python
#For RandomForestClassifier:
rf=RandomForestClassifier()
parameters={"n_estimators":[10,100,500,1000]}
clf=GridSearchCV(rf,parameters,cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
#For AdaBoostClassifier:
ada=AdaBoostClassifier()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000],"base_estimator":[lg,sv,dt],'algorithm':['SAMME']}
Gada=GridSearchCV(ada,parameters,cv=5)
Gada.fit(x,y)
Gada.best_params_
```

```python
#providing parameters
lg=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=2)
sv=SVC(kernel="rbf")
dt=DecisionTreeClassifier(random_state=9,criterion="gini")
rf=RandomForestClassifier(n_estimators=1000)
ada=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=None,
             max_features=None, max_leaf_nodes=None,
             min_impurity_decrease=0.0, min_impurity_split=None,
             min_samples_leaf=1, min_samples_split=2,
             min_weight_fraction_leaf=0.0, presort=False, random_state=6,
             splitter='best'),
 learning_rate= 1,
 n_estimators= 1000)
```

```python
models=[]
models.append(("LogisticRegression",lg))
models.append(("KNeighborsClassifier",knn))
models.append(("SVC",sv))
models.append(("DecisionTreeClassifier",dt))
models.append(("RandomForestClassifier",rf))
models.append(("AdaBoostClassifier",ada))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
Model=[]
score=[]
cvs=[]
roc_score=[]
for name,model in models:
    print('*********************',name,'***************************')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print("accuracy_score",AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring="accuracy").mean()
    print("cross_val_score=",sc)
    cvs.append(sc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,pre))
    print("\n")
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
    plt.title(name)
    print('\n\n')
```

```python
#show the results of all the models
result=pd.DataFrame({'Model':Model,'Accuracy_score':score,'Cross_Val_Score':cvs})
result
```

#since RandomForestClassifier is providing an accuracy of 76.92% and cross val score of 76.66 therefore it is selected for the model.And also it takes care of Overfitting and Underfitting problem.
