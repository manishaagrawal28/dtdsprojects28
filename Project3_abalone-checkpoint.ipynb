```python
#removing outliers
from scipy.stats import zscore
z_score=abs(zscore(abalone.iloc[:,1:]))
print(abalone.shape)
abalonedata=abalone.iloc[:,1:].loc[(z_score<3).all(axis=1)]
print(abalonedata.shape)
#final shape is changed to 4026,8 from 4176,9
```

```python
#checking for skewness
abalonedata.skew()  #skewed data present
```

```python
#removing skewness with log
import numpy as np
for col in abalonedata.columns:
    if abalonedata.skew().loc[col]>0.55:
        abalonedata[col]=np.log1p(abalonedata[col])
```

```python
#again checking for skewness
abalonedata.skew()
```

```python
#let separate input and output variables to be used for the problem
in_x=abalonedata.drop(columns=['rings'])
y=pd.DataFrame(abalonedata['rings'])
```

```python
#scaling down using standardscaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(in_x)
x=pd.DataFrame(x,columns=in_x.columns)
```

using regression algorithm for finding the output

```python
#importing algorithms and define function to find the maxr2_score corresponding to the random state

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import linear_model
def maxr2score(reg,in_x,y):
    maxr_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(in_x,y,random_state=r_state,test_size=0.30)
        reg=linear_model.LinearRegression()
        reg.fit(x_train,y_train)
        y_pred=reg.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2score corresponing to",r_state,"is:",r2_scr)
        if r2_scr>maxr_score:
            maxr_score=r2_scr
            final_r_state=r_state
    print()
    print()
    print("max r2score corresponding to ",final_r_state,"is",maxr_score) 
    return final_r_state
```

```python
#calling the function using linearRegression algorithm
reg=linear_model.LinearRegression()
r_state=maxr2score(reg,in_x,y)
```

```python
#using gridsearchcv for finding best parameters
from sklearn.model_selection import GridSearchCV
```

```python
#searching best parameters for DecisionTreeRegressor 
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
grid_param={"random_state":range(0,20)}
Gdt=GridSearchCV(dtr,param_grid=grid_param,scoring='r2',cv=10)
Gdt.fit(in_x,y)
Gdt.best_params_
```

```python
#using cross_val_score determining the bias-variance tradeoff
from sklearn.model_selection import cross_val_score
regr_score_mean=cross_val_score(reg,in_x,y,cv=5,scoring="r2").mean()
regr_score_std=cross_val_score(reg,in_x,y,cv=5,scoring="r2").std()
print("mean r2 score of linearRegression is :",regr_score_mean)
print("standard-deviation for r2 score of linearRegression is :",regr_score_std)
print('\n')

dtr_score_mean=cross_val_score(dtr,in_x,y,cv=5,scoring="r2").mean()
dtr_score_std=cross_val_score(dtr,in_x,y,cv=5,scoring="r2").std()
print("mean r2 score of DecisionTreeRegression :",dtr_score_mean)
print("stadard deviation for r2 score of DecisionTreeRegression :",dtr_score_std)
print('\n')

knr_score_mean=cross_val_score(knr,in_x,y,cv=5,scoring="r2").mean()
knr_score_std=cross_val_score(knr,in_x,y,cv=5,scoring="r2").std()
print("mean r2 score of kneighborsregression:",knr_score_mean)
print("standard deviation of r2score for kneighborsregression:",knr_score_std)
```
