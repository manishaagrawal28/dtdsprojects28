```python
#changing categorical data into numerical value

import warnings
warnings.filterwarnings("ignore")
import numpy as np
for col in churn.columns:
    if churn[col].dtypes=="object":
        from sklearn.preprocessing import LabelEncoder
        le=LabelEncoder()
        churn[col]=le.fit_transform(churn[col].astype(str))
```

```python
#removing outliers
from scipy.stats import zscore
z_score=abs(zscore(churn))
print(churn.shape)
churn=churn.loc[(z_score<3).all(axis=1)]
print(churn.shape)
#shape changed from 7043,20 to 6361,20
```

```python
#separating input and output data
#let separate input and output variables to be used for the problem
x=churn.drop(columns=['Churn'])
x=pd.DataFrame(churn.iloc[:,0:-1])
y=pd.DataFrame(churn['Churn'])
```

```python
#checking for skewness
x.skew()  #skewed data present
```

```python
#removing skewness
import numpy as np
for col in churn.columns:
    if churn.skew().loc[col]>0.55:
        churn[col]=np.log1p(churn[col])
```

```python
#again checking for skewness
churn.skew()
```

Using classification algorithm for this dataset.

```python
#splitting of the dataset
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=9,test_size=0.2,stratify=y)
```

```python
#importing algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
#creating the object
lg=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=6)
sv=SVC()
dt=DecisionTreeClassifier(random_state=6)
rf=RandomForestClassifier()
ada = AdaBoostClassifier()
```

```python
#by using gridsearch cv finding best parameters

import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
neighbors={"n_neighbors":range(1,30)}
gknn=GridSearchCV(knn,neighbors,cv=10)
gknn.fit(x,y)
gknn.best_params_
```

```python
#For DCT Classifier:

import warnings
warnings.filterwarnings('ignore')
grid_param={"random_state":range(0,20),"criterion":['gini','entropy']}
gdt=GridSearchCV(dt,param_grid=grid_param,scoring='accuracy',cv=10)
gdt.fit(x,y)
gdt.best_params_
```

```python
#For RandomForestClassifier:
rf=RandomForestClassifier()
parameters={"n_estimators":[10,100,500,1000]}
clf=GridSearchCV(rf,parameters,cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
#using GridSearchCV to find the best parameter
from sklearn.ensemble import GradientBoostingClassifier
import warnings
warnings.filterwarnings("ignore")
gbr=GradientBoostingClassifier()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000]}
clf=GridSearchCV(gbr,parameters,cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
#providing parameters
lg=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=14)
sv=SVC()
dt=DecisionTreeClassifier(random_state=11,criterion="entropy")
rf=RandomForestClassifier(n_estimators=1000)
ada=AdaBoostClassifier()
gbr=GradientBoostingClassifier(learning_rate=0.01,n_estimators=1000)
```

```python
models=[]
models.append(("LogisticRegression",lg))
models.append(("KNeighborsClassifier",knn))
models.append(("SVC",sv))
models.append(("DecisionTreeClassifier",dt))
models.append(("RandomForestClassifier",rf))
models.append(("AdaBoostClassifier",ada))
models.append(("GradientBoostClassifier",gbr))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
Model=[]
score=[]
cvs=[]
roc_score=[]
for name,model in models:
    print('*********************',name,'***************************')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print("accuracy_score",AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring="accuracy").mean()
    print("cross_val_score=",sc)
    cvs.append(sc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,pre))
    print("\n")
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
    plt.title(name)
    print('\n\n')
```

```python
#show the results of all the models
result=pd.DataFrame({'Model':Model,'Accuracy_score':score,'Cross_Val_Score':cvs})
result
```

#RandomForestClassifier is providing an accuracy of 81.22% and cross val score 79.64 and since it takes care of overfitting and underfitting problem therefore it is selected for the model.
