```python
#separating into input and output variable
df_x=df_bikes.drop(columns=['count','date'])
y=pd.DataFrame(df_bikes['count'])
```

```python
#scaling down using standardscaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
#checking for skewness
x.skew() #no skewed data present
```

```python
#finding random state value with which our linear model learns maximum
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import linear_model
from sklearn.model_selection import train_test_split
def maxr2score(reg,df_x,y):
    maxr_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(df_x,y,random_state=r_state,test_size=0.20)
        reg=linear_model.LinearRegression()
        reg.fit(x_train,y_train)
        y_pred=reg.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2score corresponing to",r_state,"is:",r2_scr)
        if r2_scr>maxr_score:
            maxr_score=r2_scr
            final_r_state=r_state
    print()
    print()
    print("max r2score corresponding to ",final_r_state,"is",maxr_score) 
    return final_r_state
```

```python
#calling the function
reg=linear_model.LinearRegression()
r_state=maxr2score(reg,df_x,y)
```

```python
#using gridsearchcv for finding the best parameters
from sklearn.model_selection import GridSearchCV
```

```python
#GridSearchCv for DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
grid_param={"random_state":range(0,20)}
Gdt=GridSearchCV(dtr,param_grid=grid_param,scoring='r2',cv=10)
Gdt.fit(df_x,y)
Gdt.best_params_
```

```python
#calling the fuction
dtr=DecisionTreeRegressor(random_state=5)
r_state=maxr2score(dtr,df_x,y)
```

```python
#using cross_val_score determining the bias-variance tradeoff

from sklearn.model_selection import cross_val_score
regr_score_mean=cross_val_score(reg,df_x,y,cv=5,scoring="r2").mean()
regr_score_std=cross_val_score(reg,df_x,y,cv=5,scoring="r2").std()
print("mean r2 score of linearRegression is :",regr_score_mean)
print("standard-deviation for r2 score of linearRegression is :",regr_score_std)
print('\n')

dtr_score_mean=cross_val_score(dtr,df_x,y,cv=5,scoring="r2").mean()
dtr_score_std=cross_val_score(dtr,df_x,y,cv=5,scoring="r2").std()
print("mean r2 score of DecisionTreeRegression :",dtr_score_mean)
print("stadard deviation for r2 score of DecisionTreeRegression :",dtr_score_std)
print('\n')

knr_score_mean=cross_val_score(knr,df_x,y,cv=5,scoring="r2").mean()
knr_score_std=cross_val_score(knr,df_x,y,cv=5,scoring="r2").std()
print("mean r2 score of kneighborsregression:",knr_score_mean)
print("standard deviation of r2score for kneighborsregression:",knr_score_std)
```
